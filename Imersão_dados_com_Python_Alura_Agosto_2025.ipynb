```python
traducao_experiencia = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['experiencia'] = df['experiencia'].replace(traducao_experiencia)
display(df['experiencia'].value_counts())
```

```python
traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Por Contrato',
    'PT': 'Meio Periodo',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(traducao_contrato)
display(df['contrato'].value_counts())
```

Aula 01

```python
traducao_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducao_tamanho_empresa)
display(df['tamanho_empresa'].value_counts())
```

```python
traducao_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(traducao_remoto)
display(df['remoto'].value_counts())
```

```python
# Exemplo de tradução para alguns dos cargos mais frequentes
traducao_cargo = {
    'Data Engineer': 'Engenheiro de Dados',
    'Data Scientist': 'Cientista de Dados',
    'Machine Learning Engineer': 'Engenheiro de Machine Learning',
    'Data Analyst': 'Analista de Dados',
    'Analytics Engineer': 'Engenheiro de Analytics',
    'Software Engineer': 'Engenheiro de Software',
    'Engineer': 'Engenheiro',
    'Data Architect': 'Arquiteto de Dados',
    'Data Modeler': 'Modelador de Dados',
    'Data Operations Analyst': 'Analista de Operações de Dados',
    'Manager': 'Gerente',
    'Analyst': 'Analista',
    'Research Scientist': 'Cientista de Pesquisa',
    'Business Analyst': 'Analista de Negócios',
    'Product Manager': 'Gerente de Produtos'
}

df['cargo'] = df['cargo'].replace(traducao_cargo)

# Exibindo os cargos mais frequentes após a tradução
display(df['cargo'].value_counts().head(10))
```

```python
salario_medio_por_experiencia = df.groupby('experiencia')['usd'].mean().reset_index()
salario_medio_por_experiencia.rename(columns={'usd': 'salario_medio'}, inplace=True)

df = df.merge(salario_medio_por_experiencia, on='experiencia', how='left')
display(df[['experiencia', 'usd', 'salario_medio']].head())
```

```python
#Importando a biblioteca NumPy
import numpy as np

#Criando um novo dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ['Eugenio', 'Liliana', 'Lia', 'Rejane', 'Luiz', 'Henrique'],
    'salario': [1000, np.nan, 2000, np.nan, 3000, 5000]
})

#Calcula a media salarial e substitui os nulos pela media e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round())

#Calcula mediana e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

#df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].fillna(df_temperaturas['temperatura'].mean().round())

#ffill substitui os nulos pelo valor antecessor
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
#bfill substitui os nulos pelo valor sucessor
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Eugenio', 'Liliana', 'Lia', 'Rejane', 'Luiz', 'Henrique'],
    'cidade': ['São Paulo', np.nan, 'Rio de Janeiro', np.nan,'Belo Horizonte', np.nan]
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

```python
import plotly.express as px
```

```python
salario_medio_por_experiencia_plotly = df_limpo.groupby('experiencia')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(salario_medio_por_experiencia_plotly,
             x='experiencia',
             y='usd',
             title='Média Salarial por Nível de Experiência',
             labels={'experiencia': 'Nível de Experiência', 'usd': 'Média Salarial Anual USD'})
fig.update_layout(
    width=800,
    height=500
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['remoto', 'quantidade']

fig = px.pie(remoto_contagem,
             values='quantidade',
             names='remoto',
             title='Distribuição de Empregos Remoto'
)
fig.update_layout(
    width=800,
    height=500
)
fig.show()
```
